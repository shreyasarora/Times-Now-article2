```python
import urllib.request
import requests
from bs4 import BeautifulSoup
import json
from urllib.request import Request,urlopen
```

```python
url='https://www.timesnownews.com/business-economy/real-estate/article/piramal-realty-gives-on-lease-1-1-lakh-sq-ft-office-space-to-pantaloons-in-mumbais-kurla/497845'
```

```python
r=Request(url,headers={'User-Agent':'Mozilla/5.0'})
#page=urlopen(r).read()
open_=urllib.request.urlopen(r)
```

```python
data=BeautifulSoup(open_,"html.parser")
print(data)
```

```python
h_data=data.find_all('h1')
for head in h_data:
    print(head.text)
f=open("article.txt",'w',encoding='UTF-8')
f.write("URL:-"+(url)+'''\n\n\n''')
for i in range(0,len(h_data)):
    f.write('''\nTITLE:\n''')
    f.write(h_data[i].get_text())
    f.write('''\n\n\n''')
```

```python
dat = data.find('div',{'class': 'heading-pod-component'})
f.write('\n\n'+"Category | Date&Time:-\n "+dat.get_text()+ '\n\n\n')
```

```python
p_data=data.find_all('p')
for para in p_data:
    print(para.text)

for i in range(0,len(p_data)):
    f.write("Content:\n")
    f.write(p_data[i].get_text())
```

```python
related_url=data.find_all('a')
f.write('''\n\n\nRelated urls:-\n''')
print(related_url)
for i in related_url:
    f.write(i['href']+'\n')  
```

```python
images=data.find_all('img')
f.write('''\n\n\nImage urls:-\n''')
print(images)
for i in images:
    f.write(i['src']+'\n')   
```
